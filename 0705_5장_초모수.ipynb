*R.square(결정 계수) - 독립 변수가 종속변수에 얼마만큼 설명해 줄 수 있는지의 가리키는 지표

y = ax + b
   X     ->      Y
독립변수        종속변수
- 새로운 X값에 대해서 모델(식)을 통해 Y값을 결정한다.
- 오차를 설명하는 R*2(R-square) =  회귀 제곱 합 / 총 제곱 합 == 1 - 잔차 제곱합 / 총 제곱 합
- 총 제곱 합 : 평균과 예측값이 얼마나 떨어져 있는지의 합
- 잔차 제곱 합 : 관측값과 예측값이 얼마나 떨어져 있는지의 합
