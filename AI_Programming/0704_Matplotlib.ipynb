```python
from datetime import timedelta
def convert_time(s):
    h, m, s = map(int, s.split(':'))
    return timedelta(hours=h, minutes=m, seconds=s)
data = pd.read_csv('https://raw.githubusercontent.com/jakevdp/marathon-data/master/marathon-data.csv', converters={'split':convert_time, 'final':convert_time})
data.head()
```

```python
data_df = pd.read_csv('https://raw.githubusercontent.com/jakevdp/marathon-data/master/marathon-data.csv')
data_df['split'] = data_df['split'].apply(convert_time)
data_df['final'] = data_df['final'].apply(convert_time)
```

```python
data_df.dtypes
```

```python
data.dtypes
```

```python
data['split_sec'] = data['split'].dt.total_seconds()
data['final_sec'] = data['final'].dt.total_seconds()
data.head()
```

```python
with sns.axes_style("whitegrid"):
    g = sns.jointplot(x='split_sec', y='final_sec', data=data, kind='reg')
    g.ax_joint.plot(np.linspace(4000, 16000),
                    np.linspace(8000, 32000), ':k', label='Some line')
```

```python
data['split_frac'] = 1 - 2 * data['split_sec'] / data['final_sec'] # 각 주자가 네거티브 스플릿 주자인지 포지티브 스플릿 주자인지의 정도를 측정하는 스플릿 정도(split fraction) 열
data.head()
```

```python
data.split_frac.hist()
```

```python
# 스플릿 정도의 분포도, 0.0은 전반부와 후반부를 같은 시간으로 완주한 주자들을 가리킴, 0 보다 작은 사람은 그 수치만큼 네거티브 스플릿으로 달린다고 볼 수 있음
sns.displot(data['split_frac'], kde=False); 
plt.axvline(0, c='k', ls='--');
```

```python
np.sum(data.split_frac < 0) # 네거티브 스플릿 주자 251 명
```

```python
sns.pairplot(data, hue='gender', kind='hist')
```

```python
g = sns.PairGrid(data, vars=['age', 'split_sec', 'final_sec', 'split_frac'],
                 hue='gender', palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```

```python
sns.kdeplot(data.split_frac[data.gender=='M'], label='men', fill=True)
sns.kdeplot(data.split_frac[data.gender=='W'], label='women', fill=True)
plt.legend()
plt.xlabel('split_frac');
# 이븐 스플릿에 가깝게 달리는 사람은 여성보다 남성이 훨씬 더 많다는 점
# 남성과 여성 사이가 일종의 이봉 분포(bimodal distribution)처럼 보임
```

```python
# 분포를 나이의 함수
# 분포 비교 : 바이올린 플롯(violin plot) 사용 좋음
sns.violinplot(x='gender', y='split_frac', data=data, hue='gender', palette=["lightblue", "lightpink"], legend=False);
```

```python
# 나이 함수로 바이올린 플롯 비교
# 배열에 각 사람의 나이를 10살 단위로 지정
data['age_dec'] = data.age.map(lambda age: 10 * (age // 10))
data.head()
```

```python
men = (data.gender == 'M')
women = (data.gender == 'W')

with sns.axes_style('whitegrid'):
    sns.violinplot(x="age_dec", y="split_frac", hue="gender", data=data,
                   split=True, inner="quartile", palette=["lightblue", "lightpink"]);
# 20대와 50대에서 남성의 스플릿 분포가 같은 연령대(또는 이 문제에 관한 한 모든 연령대)의 여성과 비교했을 때 낮은 스플릿에서 밀도가 확연히 더 높음
```

```python
g = sns.lmplot(x='final_sec', y='split_frac', col='gender', data=data, markers="*", scatter_kws=dict(color='c'))
g.map(plt.axhline, y=0.1, c='k', ls=":");
```

## plt.GridSpec: 복잡한 배치
> 서브플롯을 규칙적으로 배치하는 그리드를 넘어 여러 행과 열로 확장하려면 `plt.GridSpec()`을 사용하는 것이 좋음   
> 이 객체는 단순히 `plt.subplot()` 명령어가 인식하는 편리한 인터페이스일 뿐

```python
# 너비와 높이 여백을 지정한 두 개의 행과 세 개의 열로 이뤄진 그리드에 대한 GridSpe
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.4)

plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

```python
# 다중 히스토그램 만들 때 가장 자주 사용
# 정규 분포 데이터 만들기
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# gridspec으로 축 설정하기
fig = plt.figure(figsize=(6, 6))
gird = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(gird[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(gird[-1, 1:], yticklabels=[], sharex=main_ax)

# 메인 축에 점 산포하기
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

# 보조 축상에 히스토그램 만들기
x_hist.hist(x, 40, histtype='stepfilled', orientation='vertical', color='red')
y_hist.hist(y, 40, histtype='stepfilled', orientation='horizontal', color='blue')
y_hist.invert_xaxis()
```
